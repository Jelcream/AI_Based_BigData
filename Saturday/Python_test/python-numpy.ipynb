# numpy
 여러 항목으로 구성된 데이터를 다루려면 앞에서 배운 파이썬의 리스트나 튜플을 사용할 수 있다.
+ 다루어야 할 데이터 모두 숫자인 경우는 계산 속도를 개선하기 위해 numpy 모듈을 사용
+ Numerical Python -> numpy 다차원 배열 연산에 유리

### numpy 가 제공하는 다차원 배열 (ndarray: n dimentional array)
+ 계산을 할 때 편리하고 속도도 빠름
+ 리스트와 달리 배열에서 각 항목은 모두 같은 타입인 숫자(정수, 소수)이어야함
+ 배열을 기존 리스트로부터 만들 수 있음
+ list 만들고 여기에 3을 곱한 결과와 리스트를 배열 arr로 바꾸고 여기에 3을 곱했을 때의 결과를 비교


```python
list_1 = [1,2,10]
print(list_1*3)

arr = np.array(list_1)
print(arr * 3)
print(arr + 100)
```

+ 하나의 숫자가 배열의 항목 개수만큼 자동으로 확대되는 것을 벡터화라고 함
+ 벡터화는 곱셈 뿐 아니라 덧셈에도 적용
+ 배열도 []로 묶어서 표시

```python
# 2차원 배열
data = [[1,2,3], [4,5,6], [7,8,9]]
print(data)
print(type(data))

arr2 = np.array(data)
print(arr2)
print(type(arr2))
```

```python
# 2차원 배열의 접근
print(arr2[1,2])
print(arr2[1][2])
```

배열의 특정 위치를 인덱싱하여 얻을 수 있음
arr2의 두 번째 행의 세 번째 열의 값인 6을 얻음

### Slice
행의 일부 또는 열의 일부를 얻을 수 있음

```python
print(arr2[:2,1:])
```

## arrange
일정한 범위의 숫자를 자동으로 생성할 때 파이썬이 기본으로 제공하는 range가 있음
+ numpy에는 일정한 숫자의 범위를 제공하는 함수 arrange가 있음
+ 0~11사이의 숫자를 얻고 이를 4x3 크기의 2차원 배열로 재구성 하는 것
+ reshape 함수는 배열의 차원을 생성, 변경할 때 사용

```python
data =  np.arange(12).reshape(4,3)
print(data)
```

이 data 배열에서 항목의 값이 4보다 작은 것을 찾아서 이를 모두 0으로 대체하는 코드


```python
data[data<4] = 0
print(data)
```

data < 4 부분의 동작 결과는 True 또는 False 들로 구성된 2차원 배열을 얻게 되고 이 배열을 data[]의 인자로 다시 적용함으로써 True에 해당되는 data배열의 항목만 0으로 대체된 새로운 배열을 얻는다.

```python
print(data<4)
```

```python
print(data.mean())
```

배열 전체의 평균이 아니고 각 행 또는 각 열에 대해서 평균을 구하는 방법을 아래에 보였다.
mean() 함수의 속성값 axis를 
+ 0으로 지정하면 열을 기준으로 평균 값을 구하고, 
+ 1로 지정하면 행을 기준으로 평균값을 구함

```python
print(data.mean(axis=0))
print(data.mean(axis=1))
```
