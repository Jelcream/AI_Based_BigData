```python
fig = plt.figure()
graph = fig.add_subplot(2,1,1)
graph.plot(randn(50).cumsum(),'k--')
# cumsum 은 누적합
print(randn(50))
```

```python
fig = plt.figure()
graph = [fig.add_subplot(2,2,1),fig.add_subplot(2,2,2),fig.add_subplot(2,2,3),fig.add_subplot(2,2,4)]
color = ['k', 'g', 'r', 'b']
x = 0
for i in graph:
    i.plot(randn(50).cumsum(),color[x])
    x+=1
```
