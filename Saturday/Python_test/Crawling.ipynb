대신 id와 find()를 사용하여 원하는 항목을 찾을 수 있다.

```python
html = """
<html>
<body>
    <h1 id="title"> 파이썬으로 웹문서 읽기</h1>
    <p id="body">페이지 분석 기능</p>
    <p>페이지 정렬 </p>
</body></html>
"""
```

```python
soup = BeautifulSoup(html,'html.parser')
title = soup.find(id="title")
body = soup.find(id="body")

print(title.string)
print(body.string)
```

```python
html = """
<html>
<body>
    <ul>
    <li><a href = "http://www.naver.com">naver</a></li>
    <li><a href = "http://www.daum.com">daum</a></li>
    </ul>
</body></html>
"""
soup = BeautifulSoup(html,'html.parser')
links = soup.find_all("a")
for aa in links:
    href = aa.attrs["href"]
    text = aa.string
    print(text, "-->", href)
```

## CSS
+ CSS 선택자를 사용해서 원하는 요소를 추출할 수 있다. 아래는 h1 과 li 태그를 추출하는 코드임


```python
html= """
<html><body>
<div id="meigen">
    <h1>위키북스 도서</h1>
    <ul class="item">
        <li>게임 입문</li>
        <li>파이썬 입문</li>
        <li>웹 디자인 입문</li>
        </ul>
    </div>
</html></body>
"""
```

```python
soup = BeautifulSoup(html, 'html.parser')
h1 = soup.select_one("div#meigen > h1").string
print(h1)

li_list = soup.select("div#meigen > ul.item > li")
for li in li_list:
    print(li.string)
```

```python
url = """https://finance.naver.com/"""
res = req.urlopen(url)
soup = BeautifulSoup(res, 'html.parser')
print(soup)
price = soup.select_one("#content > div.article2 > div.section1 > div.group1 > table > tbody > tr.same.bold > td:nth-child(2)").string
print("price ==",price)
```

#content > div.article2 > div.section1 > div.group1 > table > tbody > tr.same.bold > td:nth-child(2)
